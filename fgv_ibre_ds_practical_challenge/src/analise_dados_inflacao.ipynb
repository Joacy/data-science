## Importando a planilha com os índices

```python
indices_df = pd.read_excel("../data/indices.xls")
```

```python
indices_dict = {
    '1': "IPCA",
    '2': "IGP-M",
    '3': "IPC-BR-M",
    '4': "IPA-EP-M",
    '5': "INCC-M"
}
```

```python
indices_df = indices_df[16:-1]
indices_df
```

```python
indices_df = indices_df[["Unnamed: 0", "Unnamed: 1", "Unnamed: 2", "Unnamed: 3", "Unnamed: 4", "Unnamed: 5"]]
columns = indices_df.iloc[0].tolist()
indices_df.columns = columns
```

```python
indices_df = indices_df.iloc[1:-1]
indices_df
```

```python
indices_df = indices_df.reset_index()
```

```python
indices_df = indices_df[['Data', '1', '2', '3', '4', '5']]
indices_df
```

```python
from dateutil import parser

datas = []
for data in indices_df["Data"]:
    date_time = parser.parse(data)
    datas.append(date_time)

indices_df["Data"] = datas
```

```python
jan_95 = (indices_df['Data'].dt.year >= 1995)
indices_df_Jan1995_Jun2022 = indices_df[jan_95]
indices_df_Jan1995_Jun2022
```

```python
indices_df_Jan1995_Jun2022['3'] = indices_df_Jan1995_Jun2022['3'].replace([" - "], [np.nan])
indices_df_Jan1995_Jun2022
```

```python
indices_df_Jan1995_Jun2022.columns = [["Data", indices_dict['1'], indices_dict['2'], indices_dict['3'], indices_dict['4'], indices_dict['5']]]
indices_df_Jan1995_Jun2022
```
